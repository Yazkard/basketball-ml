## Informations about dataset

### Variables

1. **TEAM:** The Division I college basketball school **(nazwa zespołu)**

2. **CONF:** The Athletic Conference in which the school participates in **(w której konferencji wystepuje zespół)**
    1. A10 = Atlantic 10, 
    2. ACC = Atlantic Coast Conference, 
    3. AE = America East, 
    4. Amer = American, 
    5. ASun = ASUN, 
    6. B10 = Big Ten, 
    7. B12 = Big 12, 
    8. BE = Big East, 
    9. BSky = Big Sky, 
    10. BSth = Big South, 
    11. BW = Big West, 
    12. CAA = Colonial Athletic Association, 
    13. CUSA = Conference USA, 
    14. Horz = Horizon League, 
    15. Ivy = Ivy League, 
    16. MAAC = Metro Atlantic Athletic Conference, 
    17. MAC = Mid-American Conference, 
    18. MEAC = Mid-Eastern Athletic Conference, 
    19. MVC = Missouri Valley Conference, 
    20. MWC = Mountain West, 
    21. NEC = Northeast Conference, 
    22. OVC = Ohio Valley Conference, 
    23. P12 = Pac-12, 
    24. Pat = Patriot League, 
    25. SB = Sun Belt, 
    26. SC = Southern Conference, 
    27. SEC = South Eastern Conference, 
    28. Slnd = Southland Conference, 
    29. Sum = Summit League, 
    30. SWAC = Southwestern Athletic Conference, 
    31. WAC = Western Athletic Conference, 
    31. WCC = West Coast Conference)

3. **G:** Number of games played **(liczba rozegranych meczów)**

4. **W:** Number of games won **(liczba wygranych meczów)**

5. **ADJOE:** Adjusted Offensive Efficiency (An estimate of the offensive efficiency (points scored per 100 possessions) a team would have against the average Division I defense) **(procentowa skuteczność ataku, którą zespół osiągnąłby przeciwko typowej obronie drużyny z Division I)**

6. **ADJDE:** Adjusted Defensive Efficiency (An estimate of the defensive efficiency (points allowed per 100 possessions) a team would have against the average Division I offense) **(procentowa skuteczność obrony, którą zespół osiągnąłby przeciwko typowemu atakowi drużyny z Division I)**

7. **BARTHAG:** Power Rating (Chance of beating an average Division I team) **(jak silna jest drużyna; szanse na pokonanie typowej drużyny Division I)**

8. **EFG_O:** [Effective Field Goal Percentage](https://en.wikipedia.org/wiki/Effective_field_goal_percentage) Shot **(ile rzutów z gry (2 lub 3 pkt))**

9. **EFG_D:** Effective Field Goal Percentage Allowed

10. **TOR:** Turnover Percentage Allowed (Turnover Rate) **(ile strat piłki)**

11. **TORD:** Turnover Percentage Committed (Steal Rate) **(ile przejęć piłki po stracie przeciwnika)**

12. **ORB:** [Offensive Rebound Percentage](https://en.wikipedia.org/wiki/Rebound_rate) **(ile zbiórek w ataku)**

13. **DRB:** [Defensive Rebound Percentage](https://en.wikipedia.org/wiki/Rebound_rate) **(ile zbiórek w obronie)**

14. **FTR:** Free Throw Rate (How often the given team shoots Free Throws) **(jak często zespół trafia rzuty osobiste (1pkt))**

15. **FTRD:** Free Throw Rate Allowed

16. **2P_O:** Two-Point Shooting Percentage **(rzuty za dwa punkty)**

17. **2P_D:** Two-Point Shooting Percentage Allowed

18. **3P_O:** Three-Point Shooting Percentage **(rzuty za trzy punkty)**

19. **3P_D:** Three-Point Shooting Percentage Allowed

20. **ADJ_T:** Adjusted Tempo (An estimate of the tempo (possessions per 40 minutes) a team would have against the team that wants to play at an average Division I tempo) **(posiadanie piłki, jakie zespół osiągnąłby w starciu z typową drużyną Division I)**

21. **WAB:** Wins Above Bubble (The bubble refers to the cut off between making the NCAA March Madness Tournament and not making it)

22. **POSTSEASON:** Round where the given team was eliminated or where their season ended
    1. R68 = First Four, 
    2. R64 = Round of 64, 
    3. R32 = Round of 32, 
    4. S16 = Sweet Sixteen, 
    5. E8 = Elite Eight, 
    6. F4 = Final Four, 
    7. 2ND = Runner-up, 
    8. Champion = Winner of the NCAA March Madness Tournament for that given year

23. **SEED:** Seed in the NCAA March Madness Tournament

24. **YEAR:** Season
